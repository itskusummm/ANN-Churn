```python
import tensorflow as tf
from tensorflow.keras.models import load_model
import pickle
import pandas as pd
import numpy as np
```

```python
# Load the trained model, scaler pickle file, and label encoder pickle file
model = load_model("my_model.h5")

# load the encoder and scaler
with open('label_encoder_gender.pkl', 'rb') as file:
    label_encoder_gender =pickle.load(file)

with open('one_hot_encoder_geo.pkl', 'rb') as file:
    one_hot_encoder_geo= pickle.load(file)

with open('scaler.pkl', 'rb') as file:
    scaler= pickle.load(file)
```

```python
# Example input data for prediction
input_data={
    'CreditScore' : 600,
    'Geography' : 'France',
    'Gender' : 'Male',
    'Age' : 40,
    'Tenure' : 3,
    'Balance' : 60000,
    'NumOfProducts' : 2,
    'HasCrCard' : 1,
    'IsActiveMember' : 1,
    'EstimatedSalary' : 50000
}
input_df=pd.DataFrame([input_data])
input_df
```

```python

geo_encoder=one_hot_encoder_geo.transform([input_df['Geography']]).toarray()
geo_encoded_df=pd.DataFrame(geo_encoder, columns=one_hot_encoder_geo.get_feature_names_out(['Geography']))
geo_encoded_df
```

```python
print(one_hot_encoder_geo.categories_[0])
print(label_encoder_gender.classes_)
```

```python
input_df=pd.concat([input_df.drop('Geography', axis=1), geo_encoded_df], axis=1)
input_df.head()
```

```python
# Encode categorical variables
input_df['Gender'] = label_encoder_gender.transform(input_df['Gender'])
input_df
```

```python
# Scale the input data
input_scaled=scaler.transform(input_df)
input_scaled
```

```python
# Predict Churn
prediction = model.predict(input_scaled)
prediction
```

```python
prediction_prob= prediction[0][0]
prediction_prob
```

```python
if prediction_prob > 0.5:
    print("The customer is likely to churn.")
else:
    print("The customer is not likely to churn.")
```
